```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
n_images = 10
kappa_list = [1, 2, 4, 6, 8, 10, 12]
fresult = './result/cactus_syn/10/'
for kappa in kappa_list:
    psnr_arr_turb_np_fname = os.path.join(fresult, 'n=10_kappa={}_psnr_arr_turb_np.npy'.format(kappa))
    psnr_arr_turb_np = np.load(psnr_arr_turb_np_fname) 
    plt.figure(0)
    plt.plot(psnr_arr_turb_np[:1000])
    plt.ylabel('PSNR turb')
    
plt.legend(['kappa=1', 'kappa=2', 'kappa=4', 'kappa=6', 'kappa=8', 'kappa=10','kappa=12'])
    
```

```python
for kappa in kappa_list:
    psnr_arr_sharp_np_fname =  os.path.join(fresult, 'n=10_kappa={}_psnr_arr_sharp_np.npy'.format(kappa))
    psnr_arr_sharp_np = np.load(psnr_arr_sharp_np_fname) 
    plt.figure(1)
    plt.plot(psnr_arr_sharp_np[:1000])
    plt.ylabel('PSNR sharp')
    
plt.legend(['kappa=1', 'kappa=2', 'kappa=4', 'kappa=6', 'kappa=8', 'kappa=10','kappa=12'])
```

```python
for kappa in kappa_list:
    ssim_arr_turb_np_fname =  os.path.join(fresult, 'n=10_kappa={}_ssim_arr_turb_np.npy'.format(kappa))
    ssim_arr_turb_np = np.load(ssim_arr_turb_np_fname) 
    plt.figure(2)
    plt.plot(ssim_arr_turb_np[:1000])
    plt.ylabel('SSIM turb')
    
plt.legend(['kappa=1', 'kappa=2', 'kappa=4', 'kappa=6', 'kappa=8', 'kappa=10','kappa=12'])
```

```python
for kappa in kappa_list:
    ssim_arr_sharp_np_fname =  os.path.join(fresult, 'n=10_kappa={}_ssim_arr_sharp_np.npy'.format(kappa))
    ssim_arr_sharp_np = np.load(ssim_arr_sharp_np_fname) 
    plt.figure(3)
    plt.plot(ssim_arr_sharp_np[:1000])
    plt.ylabel('SSIM sharp')
    
plt.legend(['kappa=1', 'kappa=2', 'kappa=4', 'kappa=6', 'kappa=8', 'kappa=10','kappa=12'])
```
